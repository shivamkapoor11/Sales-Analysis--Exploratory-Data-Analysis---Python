```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Store Sales.csv')
```

```python
df.info()
```

```python
# Changing the Datatypes from float to int
```

```python
df['Unit Price'] = df['Unit Price'].astype(int)
```

```python
df['Unit Cost'] = df['Unit Cost'].astype(int)
```

```python
df['Total Revenue'] = df['Total Revenue'].astype(int)
```

```python
df['Total Cost'] = df['Total Cost'].astype(int)
```

```python
df['Total Profit'] = df['Total Profit'].astype(int)
```

```python
df.head()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df.head()
```

# Exploratory Data Analysis using Matplotlib and Seaborn

```python
sns.set(style = 'whitegrid')
```

```python
plt.figure(figsize = (16,12))
sns.countplot(x = 'Region',data = df)
plt.title('Orders by Region', fontdict = {'fontweight':'bold', 'fontsize' : 22})
```

- This chart shows the Total Orders by different Regions. The most orders were placed in Europe and the least orders were placed in North America

```python
plt.figure(figsize = (14,10))
plt.title('Total Cost by Each Item Type (in Millions)',fontdict = {'fontweight':'bold', 'fontsize' : 18})
sns.barplot(x = 'Item Type', y = 'Total Cost', data = df, estimator = sum)
```

- The Total Cost for Office Supplies was the highest among all the Item Type sold in the store, The cost of producing Fruits 
was the lowest.

```python
plt.figure(figsize = (14,10))
plt.title('Total Revenue by Each Item Type(in Millions)', fontdict = {'fontweight':'bold', 'fontsize' : 22})
sns.barplot(x = 'Item Type', y = 'Total Revenue', data = df, estimator = sum)
```

- Revenue is the total amount of income generated by the sale of goods or services related to the company's primary operations.
- Office supplies generated the maximum revenue of all the item types, followed by houselhold goods.

```python
labels = df['Sales Channel'].value_counts().index
explode = [0,0.05]
colors = ['Green','Orange']
values = df['Sales Channel'].value_counts().values
```

```python
plt.figure(figsize = (9,9))
plt.pie(values,explode = explode,colors = colors,labels = labels,autopct = '%1.1f%%')
plt.title('Percentage of Sales Channel Selected', fontdict = {'fontweight': 'bold', 'fontsize' : 20})
```

- 52% of all the orders were placed Offline , 48% were placed online.

```python
plt.figure(figsize = (14,10))
plt.title('Total Units Sold by Each Item', fontdict = {'fontweight' : 'bold','fontsize': 18})
sns.barplot(x = 'Item Type', y = 'Units Sold',data = df, estimator = sum)
```

- Beverages was sold the most out of all the item types and the least was Fruits.

```python
plt.figure(figsize = (14,12))
plt.title('Total Profit by Each Item', fontdict = {'fontweight' : 'bold','fontsize': 16})
sns.barplot(x = 'Item Type', y = 'Total Profit', data = df, estimator = sum)
```

- Profits is the amount of income that remains after accounting for all expenses, debts, additional income streams and operating costs.
- Cosmetics yielded the maximum profits out of all the Items and least profit was for Fruits

```python
plt.figure(figsize = (8,8))
df['Order Priority'].value_counts().plot(kind = 'bar')
```

- Most of the orders priority was Low.

```python
plt.figure(figsize = (10,10))
sns.distplot(df['Total Revenue'])
```

- Distplot shows the average disribution of the numerical column. In this case we chose Total Revenue. It can be seen most of
the profit was between 0-1400000.

```python
plt.figure(figsize = (12,8))
sns.boxplot(x = 'Region',y = 'Units Sold', data = df)
```

```python
plt.figure(figsize = (14,10))
plt.title('Total Profit by Region', fontdict = {'fontweight' : 'bold', 'fontsize' : 16})
sns.barplot(x = 'Region', y = 'Total Profit', data = df, estimator = sum )
```

- Europe had the maximum profits out of all the regions. The least profits were for the North America Region.

```python
sns.pairplot(df)
```

# Interpretations

- Most Orders were from Europe.
- Total Profit increased as the total units sold increased. There is a direct relation of Profit and Units sold.
- People preferred to buy offline than order online. 52% in favour of buying it offline.
- The most units sold was for Beverages, however, the Total Profits and the Total Revenue generated from it was the least of all items.
- The Average profit was between 0-1400000 for all the items sold.
- As Most Orders were from Europe, it can be seen that the Highest Profit was from Europe.
